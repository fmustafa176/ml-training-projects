```python
# imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import kagglehub
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# load dataset

path = kagglehub.dataset_download("gurdit559/canada-per-capita-income-single-variable-data-set")
df=pd.read_csv(path + '/canada_per_capita_income.csv')
df.head()
```

```python
x = df.iloc[:, 0:1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
reg.predict([[2020]])
```

```python
# plot

plt.scatter(x_train, y_train)
plt.plot(x_train, reg.predict(x_train), color='red')
plt.scatter([2020], reg.predict([[2020]]), color='green', label="Predicted points")

plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.title('Canada Per Capita Income')
plt.show()
```
